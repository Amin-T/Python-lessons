```
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

%matplotlib inline
import numpy as np
```

### **matplotlib**

```
x = np.arange(0,10)
y = x**2
```

```
plt.plot(x, y, 'r')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```
fig = plt.figure()
plt.plot(x, y, 'r')
plt.xlabel('X')
plt.ylabel('Y')
```

```
fig.savefig('fig1.jpg', dpi=100)
```

## **pandas**

```
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv'
wine_df = pd.read_csv(url, sep=';')
wine_df.iloc[:5, -5:]
```

```
plot = wine_df.plot(x='alcohol', y='pH', kind = 'scatter', alpha=0.5, figsize=(10, 6))
```

```
fig = plot.get_figure()
```

```
fig.savefig('fig2.jpg')
```

## **seaborn**

```
x = np.arange(0,2*np.pi, np.pi/16)
y = np.sin(x)
```

```

plot = sns.lineplot(x=x, y=y)

plt.title('sine')
plt.xlabel('X')
plt.ylabel('Y')
```

```
plot.figure.savefig('fig3.jpg')
```
