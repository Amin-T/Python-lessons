## **Merging DataFrames**

### **Concatenate**

```
df1 = pd.DataFrame({'A': ['1Aa', '1Ab', '1Ac', '1Ad'],
                    'B': ['1Ba', '1Bb', '1Bc', '1Bd'],
                    'C': ['1Ca', '1Cb', '1Cb', '1Cd'],
                    'D': ['1Da', '1Db', '1Dc', '1Dd']}, index=['a', 'b', 'c', 'd'])

df1
```

```
df2 = pd.DataFrame({'A': ['2Ae', '2Af', '2Ag', '2Ah'],
                    'B': ['2Be', '2Bf', '2Bg', '2Bh'],
                    'C': ['2Ce', '2Cf', '2Cg', '2Ch'],
                    'D': ['2De', '2Df', '2Dg', '2Dh']}, index=['e', 'f', 'g', 'h'])

df2
```

```
df3 = pd.DataFrame({'A': ['3Aa', '3Ab', '3Ae', '3Af'],
                    'B': ['3Ba', '3Bb', '3Be', '3Bf'],
                    'E': ['3Ca', '3Cb', '3Ce', '3Cf'],
                    'F': ['3Da', '3Db', '3De', '3Df']}, index=['a', 'b', 'e', 'f'])

df3
```

```
pd.concat([df1,df2,df3], axis=0)
```

```
pd.concat([df1,df2,df3], axis=1)
```

### **Join**

```
left_df = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                        'B': ['B0', 'B1', 'B2']},
                       index=['K0', 'K1', 'K2']) 

left_df
```

```
right_df = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                         'D': ['D0', 'D2', 'D3']},
                        index=['K0', 'K2', 'K3'])

right_df
```

```
left_df.join(right_df)
```

```
left_df.join(right_df, how='outer')
```

### **Merge**

```
df1 = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                    'key2': ['K0', 'K1', 'K0', 'K1'],
                    'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3']})
    
df1
```

```
df2 = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                    'key2': ['K0', 'K0', 'K0', 'K0'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']})

df2
```

```
pd.merge(left=df1, right=df2)
```

```
pd.merge(left=df1, right=df2, how='inner', on=['key1', 'key2'])
```

```
pd.merge(left=df1, right=df2, how='left', on='key1')
```

```
pd.merge(left=df1, right=df2, how='right', on='key2')
```

```
pd.merge(left=df1, right=df2, left_on='key2', right_on='key1')
```
